#**1. FIXING MISSING VALUES**

##-- emp column

```python
print(f'the total missing values in emp are {np.isnan(emp).sum()}')
#we are going to replace it with median values because median has no affect on most outlying values
median_emp=np.nanmedian(emp)
print(f"median value of emp is {median_emp}")

missing_emp=np.where(np.isnan(emp))[0]
print("missing emp:",missing_emp)
emp[missing_emp]=median_emp
emp
```

```python
print(f'the total missing values in emp are {np.isnan(emp).sum()}')
print(f'the max value in emp are {np.max(emp)}')
```

##--intrest column

```python
print(f'the total missing values in intrest are {np.isnan(intrest).sum()}')
#we are going to replace it with median values because median has no affect on most outlying values
median_intrest=np.nanmedian(intrest)
print(f"median value of intrest is {median_intrest}")

missing_intrest=np.where(np.isnan(intrest))[0]
print("missing intrest:",missing_intrest)
intrest[missing_intrest]=median_intrest
intrest
```

```python
print(f'the total missing values in intrest are {np.isnan(intrest).sum()}')
print(f'the max value in intrest are {np.max(intrest)}')
```

# **2) FIXING OUTLIERS**

```python
def outliers(column):
  q1=np.percentile(column,25)
  q3=np.percentile(column,75)

  iqr=q3-q1
  lower=q1-(iqr*1.5)
  upper=q3+(iqr*1.5)

  outliers=np.where((column<lower)|(column>upper))[0]
  print("lower:",lower)
  print("upper:",upper)
  return outliers
def lower(column):
  q1=np.percentile(column,25)
  q3=np.percentile(column,75)

  iqr=q3-q1
  lower=q1-(iqr*1.5)
  return lower
def upper(column):
  q1=np.percentile(column,25)
  q3=np.percentile(column,75)

  iqr=q3-q1
  upper=q3+(iqr*1.5)
  return upper
```

###clean data for age

```python
print("outliers in age(index):",outliers(age))
```

```python
#but the age limit of giving loan is between 18 and 70
out_age=np.where((age<18)|(age>70))[0]
print("Outlier count:", out_age.shape)
print("Outlier age:", age[out_age])

#now valid ages
valid_age = age[(age >= 18) & (age <= 70)]

#median on ages
median = np.median(valid_age)
print("median of ages is:",median)
# replace invalid ages
age[out_age]=median
age
```

###clean data for income

```python
outliers(income)
print(f"max income is {np.max(income)}")
print(f"the 99th percentile of income is {np.round(np.percentile(income,99))}")
print(f"count if income above 99th percentile {np.sum(income>225200)}")
```

```python
#so we have 99th percentile as 225200 and max income as 6000000 which can be a potential outlier but the count is low so we can replace or cap it with nearest percentile value
out_income = np.where((income < 1000) | (income > 225200))[0]
print("Outlier count:", out_income.shape)
print("Outlier income(inr):", income[np.where(income>225200)][:10])

# get valid income values
valid_income = income[(income >= 1000) & (income <= 225200)]

# correct income
correct_income = 225200
print("99th percent corrected income:", correct_income)

# replace outliers with median
income = np.where((income < 1000) | (income > 225200), correct_income, income)
income
```

```python
print(f"count if income above 99th percentile is {np.sum(income>225200)}")
```
