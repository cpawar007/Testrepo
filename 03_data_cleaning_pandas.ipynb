```python
df=pd.read_csv("cleaned_numpy_data.csv")
```

```python
df=pd.DataFrame(df)
```

```python
df=df.rename(columns={'0':'age','1':'income','2':'ownership','3':'emp','4':'intend','5':'grade','6':'amount','7':'intrest','8':'status','9':'percent','10':'default','11':'history'})
```

```python
df.head()
```

```python
#checking types and unique values
print(df.dtypes)
print(df['intend'].unique())
print(df['grade'].unique())
print(df['ownership'].unique())
print(df['emp'].unique())
print(df['default'].unique())
```

```python
#check for missing values
df.select_dtypes(include='object').isna().sum()
```

```python
#have to check for potential outliers in income
print(f"max income:{df['income'].max()} ; min income:{df['income'].min()}")
print(f"mean of income:{df['income'].mean()} ; median of income:{df['income'].median()}")
#we can see mean>median hence its right skewed so outliers exist
```

```python
q1= df['income'].quantile(0.25)
q3= df['income'].quantile(0.75)
iqr= q3-q1
lower_bound= q1-1.5*iqr
upper_bound= q3+1.5*iqr
print(f"lower bound:{lower_bound} ; upper bound:{upper_bound}")
```

```python
df['income'].quantile(0.99)
#we have 99th place of data which can be used to deal with outlier as it is more higher than upper bound
```

```python
df['income']=df['income'].clip(upper=df['income'].quantile(0.99))
```

```python
print(f"max income:{df['income'].max()} ; min income:{df['income'].min()}")
print(f"mean of income:{df['income'].mean()} ; median of income:{df['income'].median()}")
```

```python
df.to_csv('cleaned_pandas_dataaa.csv', index=False)
```
